```python
import matplotlib.pyplot as plt

#1.製作figure
fig = plt.figure()

#2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1,1,1)

#3.將圖的數據儲存在axes中
X=[0,1,2]
Y=[3,4,5]
ax_1.plot(X,Y)

#4.顯示圖
plt.show()
```

```python
fig = plt.figure()
ax_1 = fig.add_subplot(1,3,1)
ax_1.set_title("axes1")

ax_2 = fig.add_subplot(1,3,2)
ax_2.set_title("axes2")

ax_3 = fig.add_subplot(1,3,3)
ax_3.set_title("axes3")
```

```python
student1 = {'chinese':80,'english':70,'math':60,'history':50,'discover':70}
student2 = {'chinese':70,'english':60,'math':75,'history':90,'discover':85}

subjects = list(student1.keys())
value1 = list(student1.values())
value2 = list(student2.values())

figure = plt.figure(figsize=(8.0,5.0), dpi=250.0, facecolor='#cccccc')
ax = figure.add_subplot()
ax.plot(subjects, value1)

ax.plot(subjects, value2)

plt.show()
```
